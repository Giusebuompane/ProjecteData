## Sobre la Encuesta de Competencias Financieras (ECF)
***
1. ¿Qué mide esta fuente de datos? <br> <ul>
<li>Objetivo principal: Mide el nivel de conocimiento y comportamiento financiero de la población adulta en España y sus actitudes ante la planificación y toma de decisiones económicas.<br>
<li>Información capturada: Hábitos de ahorro, planificación financiera, uso de productos financieros, conocimientos sobre inflación, interés compuesto, riesgo y diversificación.<br></ul>

2. ¿Qué variables o campos contiene el conjunto de datos?<br><ul>
<li>Variables principales: Nivel de estudios, situación laboral, ingresos, edad, género, región, productos financieros utilizados, conocimientos teóricos (preguntas tipo test), actitudes financieras (por ejemplo, tolerancia al riesgo).<br>
<li>Conceptos clave:<br>
<li>Literatura financiera: comprende conceptos como interés compuesto, inflación, riesgo y diversificación.<br>
<li>Comportamiento financiero: incluye planificación, ahorro y uso de productos.<br>
<li>Segmentación: Los datos están desagregados por grupo demográfico (edad, género, educación), situación económica, y ámbito territorial (comunidades autónomas).<br>
<li>Tipo de variables: Numéricas (por ejemplo, ingresos), categóricas (nivel educativo, género), ordinales (frecuencia de ahorro), y algunas temporales (edad o referencias temporales a eventos financieros).<br></ul>

3. ¿Qué cobertura temporal tienen los datos?<br><ul>
<li>Rango temporal: Encuesta realizada entre noviembre de 2020 y febrero de 2021.<br>
<li>Frecuencia: No es una encuesta anual; la anterior fue en 2016. Es periódica pero no con frecuencia fija.<br>
<li>Análisis de tendencias: Puede compararse con la edición de 2016 para observar evolución o cambios en conocimientos y hábitos financieros.<br></ul>

4. ¿Qué nivel de detalle geográfico tienen los datos?<br><ul>
<li>	Desagregación territorial: Los resultados están disponibles a nivel de comunidades autónomas.<br>
<li>Comparación regional: Sí, es posible hacer comparaciones entre regiones en aspectos como conocimientos financieros medios o acceso a productos financieros.<br></ul>

5. ¿Qué limitaciones o advertencias tienen los datos?<ul>
<li>Datos estimados o limitaciones: Aunque es una encuesta representativa, se basa en autodeclaración de los participantes, lo que puede introducir sesgos de percepción o memoria.<br>
<li>Advertencias metodológicas:<br>
<li>Diferencias entre ediciones de la encuesta pueden dificultar comparaciones directas.<br>
<li>Algunos colectivos pueden estar infrarrepresentados.<br>
<li>Notas técnicas: El documento incluye un anexo metodológico detallado con explicaciones sobre el diseño de la muestra, márgenes de error y ponderaciones utilizadas.<br></ul>

### Gráfico de barras comparativo 2016-2021: Porcentaje de individuos que posee vehículos de ahorro o deudas, por grupo de edad
  *En 2016, «algún vehículo de ahorro» incluye cuentas de ahorro, planes de pensiones, fondos de inversión, acciones y activos de renta fija. En 2021 incluye también criptomonedas*

```python
edades = ['18-34', '35-44', '45-54', '55-64', '65-79']
veh_ahorro_2016 = [28, 47, 52, 52, 41]
veh_ahorro_2021 = [29, 41, 46, 49, 41]

deudas_2016 = [23, 63, 56, 41, 20]
deudas_2021 = [24, 62, 60, 42, 22]

x = np.arange(len(edades))
width = 0.35

colors = ["#2CCED1", "#5F6F81"]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

ax1.bar(x - width/2, veh_ahorro_2016, width, label='2016', color=colors[0])
ax1.bar(x + width/2, veh_ahorro_2021, width, label='2021', color=colors[1])
ax1.set_ylabel('% de individuos')
ax1.set_title('1.a Algún vehículo de ahorro')
ax1.legend()

for i in range(len(x)):
    ax1.text(x[i] - width/2, veh_ahorro_2016[i] + 1, str(veh_ahorro_2016[i]), ha='center', fontsize=9)
    ax1.text(x[i] + width/2, veh_ahorro_2021[i] + 1, str(veh_ahorro_2021[i]), ha='center', fontsize=9)

ax2.bar(x - width/2, deudas_2016, width, label='2016', color=colors[0])
ax2.bar(x + width/2, deudas_2021, width, label='2021', color=colors[1])
ax2.set_ylabel('% de individuos')
ax2.set_title('1.b Alguna hipoteca o préstamo')
ax2.set_xticks(x)
ax2.set_xticklabels(edades)
ax2.legend()

for i in range(len(x)):
    ax2.text(x[i] - width/2, deudas_2016[i] + 1, str(deudas_2016[i]), ha='center', fontsize=9)
    ax2.text(x[i] + width/2, deudas_2021[i] + 1, str(deudas_2021[i]), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

**1a: Algún vehículo de ahorro**<br>

En ambos años (2016 y 2021), el grupo de edad 45-64 años presenta los mayores niveles de tenencia de productos de ahorro, aunque con una ligera caída en 2021.<br>

El grupo más joven (18-34) sigue mostrando un nivel muy bajo de ahorro, con apenas un 29%.<br>

**1b: Hipoteca o préstamo**<br>
El porcentaje de personas con deuda (hipoteca o préstamo) se mantiene relativamente estable en todas las edades entre los dos años.<br>

El grupo de 45-54 años muestra el mayor crecimiento en deudas (+4 puntos), lo que podría estar ligado a responsabilidades familiares o vivienda.<br>

**Recomendación estratégica**<br>
*Segmento prioritario para ahorro:*<br>
Centrar campañas de productos de ahorro en los grupos de edad 35-64 años, donde ya existe familiaridad con estos productos y capacidad de inversión.<br>

*Educación financiera para jóvenes:*<br>
El bajo conocimiento y uso de vehículos de ahorro en el grupo de 18-34 años sugiere la necesidad de campañas educativas y productos adaptados (simples, digitales, con aportaciones flexibles).<br>

*Seguimiento de la deuda en edades medias:*<br>
Ofrecer servicios de consolidación de deudas, asesoría financiera y seguros asociados en el grupo de 45-54 años, donde coinciden altos niveles de ahorro y deuda.<br>

### Jóvenes 18-34 años: Buscan productos sencillos y sin comisión, preferentemente con gestión digital, flexibilidad e incentivos como descuentos.<br>
Cuentas: Los bancos ofrecen tarjetas específicas hasta los 30 años que cumplen con estos requisitos, ofreciendo descuentos en ocio y viajes y programas(BBVA, Caixabank) vinculados a la cuenta como becas, préstamos para educación y formación en emprendimiento(Santander).<br>
Funciones digitales: estas cuentas preferidas por jóvenes incluyen herramientas modernas como pago móvil(google pay, Bizum), permiten activar redondeo de ahorro, ofrecen un chatbot para responder con inmediatez las 24 horas del día. Descuentos a tiendas online y sorteos.<br>
Tarjetas de crédito y débito: Las tarjetas de débito vienen incluidas en todas las cuentas, las hay muy adaptadas al uso digital como la Tarjeta Aqua de BBVA que no tiene números impresos y genera un cvv distinto en cada compra. <br>
Los bancos suelen ofrecer tarjetas de crédito básicas a jóvenes trabajadores, a veces con el primer año gratuito o límites reducidos para un uso responsable. Suelen incluir un programa de puntos.<br>
Productos de ahorro:Conscientes de que muchos jóvenes tienen ingresos limitados, los bancos proporcionan herramientas para fomentar el hábito de ahorrar pequeñas cantidades. Por ejemplo, la ya mencionada función de reglas de ahorro de BBVA redondea cada compra con tarjeta y guarda el excedente en una “hucha” virtual.<br>Otras apps permiten establecer metas de ahorro automáticas, apartando dinero periódicamente de la cuenta corriente. <br>Esto se complementa con educación financiera sencilla en las webs bancarias, para enseñar a invertir o ahorrar a quienes quizá se incorporan al sistema financiero por primera vez.<br>
Productos de inversión: Los jóvenes muestran cada vez más interés en invertir de forma sencilla, aprovechando las plataformas FinTech.<br> En España operan robo-advisors (gestores automatizados) que, mediante carteras indexadas de fondos, permiten invertir desde importes muy bajos (algunos desde 50 €) con comisiones.Estos productos suelen perfilar el riesgo mediante cuestionarios, lo que atrae a los menores de 35 por su comodidad y educación implícita<br>
Además, se enfatiza la inversión sostenible (ESG) porque suele resonar con los valores de esta generación. <br>Recientemente, Bankinter incluso abrió la posibilidad de contratar carteras de fondos para niños y adolescentes (a nombre de los padres como tutores), fomentando el ahorro/inversión temprana para fines futuros (estudios, coche, etc.).<br> Esto refleja una estrategia de largo plazo: captar a los clientes desde jóvenes con herramientas de inversión fáciles, que crezcan con ellos.<br>
Plataformas de trading y criptomonedas: Si bien no son productos bancarios tradicionales, es notable que muchos jóvenes utilizan apps como Revolut, eToro, Binance, Robinhood (no españolas) para invertir en acciones, criptomonedas u otros activos. <br>Los bancos españoles han respondido integrando ciertos servicios: por ejemplo, BBVA ofrece compra/venta de criptomonedas a través de su filial en Suiza (aunque aún muy limitado), y algunas entidades tienen brokers online con tarifas bajas para bolsa (ej. ClickTrade de Santander o Broker Naranja de ING).<br> No obstante, los productos más recomendados para este segmento siguen siendo los fondos diversificados y la inversión automática, por su sencillez y menor riesgo relativo.<br>

Seguros<br>
Seguros de salud y vida: En la franja 18-35, la contratación de seguros personales es menor (muchos aún dependen de coberturas familiares o la sanidad pública).<br> Aun así, las aseguradoras bancarias empiezan a ofrecer planes adaptados.<br> Por ejemplo, algunas entidades promueven seguros de salud digital para jóvenes a precios reducidos (coberturas médicas básicas con servicios online). <br>También existen seguros de vida hipotecarios pensados para jóvenes compradores de primera vivienda, pero fuera de esas situaciones no es habitual que un cliente de 25-30 contrate un seguro de vida.<br>
Seguros de hogar y auto: Estos sí entran en juego conforme los jóvenes adquieren coche o alquilan/compran vivienda. <br>Los bancos aprovechan para ofrecer seguros de coche con descuentos para jóvenes (ej. bonificaciones por buena conducción o cursos de conducción segura) a través de sus aseguradoras asociadas, aunque las primas para menores de 25 suelen ser altas por riesgo estadístico<br>
Seguros de viaje y gadgets: Dado que este público viaja con frecuencia y usa intensamente dispositivos electrónicos, algunos bancos incluyen seguros de viaje gratuitos asociados a las tarjetas jóvenes (cobertura de accidentes o equipajes en viajes comprados con la tarjeta) para añadir valor.<br> Asimismo, productos novedosos como el seguro de móvil han aparecido en catálogos bancarios .<br> En general, la estrategia con seguros para jóvenes es combinar ofertas sencillas y baratas y vincularlas a otro producto principal (p.ej., un seguro de protección de compras ligado a la tarjeta, seguro de asistencia en carretera con la cuenta nómina joven, etc.), de modo que se perciba como un extra de seguridad más que como un coste adicional.<br>
Financiación: <br>
Préstamos para estudios y formación: Un producto clave para el tramo 18-35 son los préstamos destinados a financiar estudios superiores, másters o cursos. Los bancos españoles (en ocasiones a través de filiales sociales) ofrecen préstamos educativos con condiciones especiales. <br>Estas financiaciones suelen tener tipos de interés reducidos y plazos largos, reconociendo el potencial de ingresos futuros del joven. Incluso entidades públicas y bancarias colaboran para dar microcréditos a jóvenes emprendedores o investigadores.<br>
Créditos personales y consumo: Más allá de estudios, los jóvenes pueden necesitar financiar la compra de un vehículo, mobiliario o un proyecto personal.<br> Los bancos ofrecen préstamos personales estándar también a este colectivo, aunque evalúan su solvencia (ingresos estables, contrato laboral) que a veces es limitada en los veintitantos.<br> Algunos han creado campañas específicas: por ejemplo, préstamos preconcedidos pequeños en la app (2000-3000 €) para clientes jóvenes con buena salud financiera, para viajes o imprevistos. Estos préstamos suelen tramitarse 100% online en minutos si el cliente ya es conocido por el banco<br>
Dada la dificultad de los jóvenes para ahorrar el típico 20% de entrada de una casa, varios bancos han desarrollado la Hipoteca Joven con financiación superior al estándar(Banco Santander). A cambio requiere garantías adicionales como avales. <br>

Productos diseñados a su medida<br>
