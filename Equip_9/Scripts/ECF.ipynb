## Sobre la Encuesta de Competencias Financieras (ECF)
***
1. ¿Qué mide esta fuente de datos? <br> <ul>
<li>Objetivo principal: Mide el nivel de conocimiento y comportamiento financiero de la población adulta en España y sus actitudes ante la planificación y toma de decisiones económicas.<br>
<li>Información capturada: Hábitos de ahorro, planificación financiera, uso de productos financieros, conocimientos sobre inflación, interés compuesto, riesgo y diversificación.<br></ul>

2. ¿Qué variables o campos contiene el conjunto de datos?<br><ul>
<li>Variables principales: Nivel de estudios, situación laboral, ingresos, edad, género, región, productos financieros utilizados, conocimientos teóricos (preguntas tipo test), actitudes financieras (por ejemplo, tolerancia al riesgo).<br>
<li>Conceptos clave:<br>
<li>Literatura financiera: comprende conceptos como interés compuesto, inflación, riesgo y diversificación.<br>
<li>Comportamiento financiero: incluye planificación, ahorro y uso de productos.<br>
<li>Segmentación: Los datos están desagregados por grupo demográfico (edad, género, educación), situación económica, y ámbito territorial (comunidades autónomas).<br>
<li>Tipo de variables: Numéricas (por ejemplo, ingresos), categóricas (nivel educativo, género), ordinales (frecuencia de ahorro), y algunas temporales (edad o referencias temporales a eventos financieros).<br></ul>

3. ¿Qué cobertura temporal tienen los datos?<br><ul>
<li>Rango temporal: Encuesta realizada entre noviembre de 2020 y febrero de 2021.<br>
<li>Frecuencia: No es una encuesta anual; la anterior fue en 2016. Es periódica pero no con frecuencia fija.<br>
<li>Análisis de tendencias: Puede compararse con la edición de 2016 para observar evolución o cambios en conocimientos y hábitos financieros.<br></ul>

4. ¿Qué nivel de detalle geográfico tienen los datos?<br><ul>
<li>	Desagregación territorial: Los resultados están disponibles a nivel de comunidades autónomas.<br>
<li>Comparación regional: Sí, es posible hacer comparaciones entre regiones en aspectos como conocimientos financieros medios o acceso a productos financieros.<br></ul>

5. ¿Qué limitaciones o advertencias tienen los datos?<ul>
<li>Datos estimados o limitaciones: Aunque es una encuesta representativa, se basa en autodeclaración de los participantes, lo que puede introducir sesgos de percepción o memoria.<br>
<li>Advertencias metodológicas:<br>
<li>Diferencias entre ediciones de la encuesta pueden dificultar comparaciones directas.<br>
<li>Algunos colectivos pueden estar infrarrepresentados.<br>
<li>Notas técnicas: El documento incluye un anexo metodológico detallado con explicaciones sobre el diseño de la muestra, márgenes de error y ponderaciones utilizadas.<br></ul>

 ### Porcentaje de individuos que ha oído hablar de… 
 Comparativa entre 2016 y 2021

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos del gráfico extraídos
data = {
    'Producto financiero': ['Cuenta de ahorro', 'Renta fija'],
    '2016': [73, 78],
    '2021': [60, 72]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
bar_width = 0.35
x = range(len(df))

plt.bar([p - bar_width/2 for p in x], df['2016'], width=bar_width, label='2016', color='#2CCED1')
plt.bar([p + bar_width/2 for p in x], df['2021'], width=bar_width, label='2021', color='#5F6F81')

# Etiquetas y formato
plt.xticks(ticks=x, labels=df['Producto financiero'])
plt.ylabel('% de individuos')
plt.title('Porcentaje de individuos que ha oído hablar de...')
plt.legend()
plt.tight_layout()

plt.show()
```
