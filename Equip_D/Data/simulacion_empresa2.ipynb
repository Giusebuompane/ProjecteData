## Diccionario del Dataset
### Identificación del alojamiento:
-apartment_id: Identificador único del alojamiento.<br>
-md5: Código hash único asociado al alojamiento.<br>
-name: Nombre del alojamiento.<br>
-description: Descripción detallada del alojamiento.<br>
-url: url del alojamiento turístico en alquiler.<br>

### Información del anfitrión y vecindario
-host_id: Identificador único del anfitrión (dueño o gestor del alojamiento).<br>
-Ubicación<br>
-neighborhood_overview: Descripción general del vecindario.<br>
-neighbourhood_name: Nombre del vecindario donde se ubica el alojamiento.<br>
-neighbourhood_district: Distrito o área más amplia a la que pertenece el vecindario.<br>
-latitude y longitude: Coordenadas geográficas del alojamiento.<br>

### Características del alojamiento
-room_type: Tipo de alojamiento (ejemplo: "Entire home/apt", "Private room").<br>
-accommodates: Número de personas que puede alojar.<br>
-bathrooms: Número de baños disponibles.<br>
-bedrooms: Número de habitaciones disponibles.<br>
-beds: Número de camas disponibles.<br>
-amenities_list: Lista de comodidades y servicios incluidos (ejemplo: WiFi, cocina, piscina).<br>

### Disponibilidad y precios
-price: Precio por noche en la moneda local.<br>
-minimum_nights: Número mínimo de noches requeridas para reservar.<br>
-maximum_nights: Número máximo de noches permitidas por reserva.<br>
-has_availability: Indica si el alojamiento está disponible para reservas (True o False).<br>
-availability_30, availability_60, availability_90, availability_365: Número de noches disponibles en los próximos 30, 60, 90 y 365 días, respectivamente.<br>

### Reseñas y puntuaciones
-number_of_reviews: Número total de reseñas recibidas.<br>
-first_review_date y last_review_date: Fecha de la primera y última reseña recibida.<br>
-review_scores_rating: Puntuación de 0 a 100 dada por los usuarios del alojamiento turístico.<br>
-review_scores_accuracy: Puntuación de 0 a 10 sobre la precisión de los detalles del alojamiento turístico.<br>
-review_scores_cleanliness: Puntuación de 0 a 10 sobre las condiciones higiénicas del alojamiento turístico..<br>
-review_scores_checkin: Puntuación de 0 a 10 sobre el momento de entrada al alojamiento turístico.<br>
-review_scores_communication: Puntuación de 0 a 10 sobre la comunicación del propietario con los usuarios del alojamiento turístico.<br>
-review_scores_location: Puntuación de 0 a 10 sobre la zona donde se encuentra el alojamiento turístico..<br>
-review_scores_value: Puntuación de 0 a 10 sobre la evaluación del alojamiento turístico.<br>
-reviews_per_month: Número de comentarios que recibe el alojamiento turístico al mes.<br>

### Regulación y reservas
-license: Número de licencia turística.<br>
-is_instant_bookable: Si puede reservarse sin necesidad de revisión por parte del propietario..<br>
-insert_date: Fecha de extracción de la información.<br>

### Ubicación general y fecha de registro
-country: País donde se ubica el alojamiento.<br>
-city: Ciudad donde se ubica el alojamiento.<br>

# 2do Paso: Mostrar y modificar datos

```python
# Hacemos un sumatorio de los nulos
print(Tourist_Accommodation.isnull().sum())  
```

```python
# Nos aseguramos de que no haya valores nulos (NaN).
Tourist_Accommodation['bathrooms'] = pd.to_numeric(Tourist_Accommodation['bathrooms'], errors='coerce')
Tourist_Accommodation["bedrooms"] = pd.to_numeric(Tourist_Accommodation["bedrooms"], errors='coerce')
Tourist_Accommodation["first_review_date"] = pd.to_datetime(
    Tourist_Accommodation["first_review_date"], errors="coerce"
)
Tourist_Accommodation["last_review_date"] = pd.to_datetime(
    Tourist_Accommodation["last_review_date"], errors="coerce"
)
Tourist_Accommodation["insert_date"] = pd.to_datetime(
    Tourist_Accommodation["insert_date"], errors="coerce"
)
print(Tourist_Accommodation["insert_date"].isna().sum())
```

```python
# Procedemos con las imputaciones

# Columnas numéricas mejor con mediana ya que los outliers nos pueden afectar. Por ejemplo puede haber casas con 50 baños, pero no es lo habitual.
Tourist_Accommodation['bathrooms'].fillna(Tourist_Accommodation['bathrooms'].median(), inplace=True)
Tourist_Accommodation['bathrooms'] = Tourist_Accommodation['bathrooms'].astype('int')
Tourist_Accommodation["bedrooms"].fillna(Tourist_Accommodation["bedrooms"].median(), inplace=True)
Tourist_Accommodation["beds"].fillna(Tourist_Accommodation["beds"].median(), inplace=True)
Tourist_Accommodation["price"].fillna(Tourist_Accommodation["price"].median(), inplace=True)
Tourist_Accommodation["minimum_nights"].fillna(Tourist_Accommodation["minimum_nights"].median(), inplace=True)
Tourist_Accommodation["maximum_nights"].fillna(Tourist_Accommodation["maximum_nights"].median(), inplace=True)

# Columnas de disponibilidad
Tourist_Accommodation["availability_30"].fillna(Tourist_Accommodation["availability_30"].mean(), inplace=True)
Tourist_Accommodation["availability_60"].fillna(Tourist_Accommodation["availability_60"].mean(), inplace=True)
Tourist_Accommodation["availability_90"].fillna(Tourist_Accommodation["availability_90"].mean(), inplace=True)
Tourist_Accommodation["availability_365"].fillna(Tourist_Accommodation["availability_365"].mean(), inplace=True)

# Columnas categóricas
Tourist_Accommodation["neighbourhood_district"].fillna("Desconocido", inplace=True)
Tourist_Accommodation["has_availability"].fillna("unknown", inplace=True)

# Columnas de texto
Tourist_Accommodation["name"].fillna("Desconocido", inplace=True)
Tourist_Accommodation["description"].fillna("No disponible", inplace=True)
Tourist_Accommodation['amenities_list'] = Tourist_Accommodation['amenities_list'].fillna("Sin información")

# columnas de fecha. Es mejor la mediana para fechas, ya que la mediana toma la fecha central de la distribución, lo que evita el impacto de valores extremos.Es más probable que la fecha resultante esté dentro del rango normal de la columna.
Tourist_Accommodation["first_review_date"].fillna(Tourist_Accommodation["first_review_date"].median(), inplace=True)
Tourist_Accommodation["last_review_date"].fillna(Tourist_Accommodation["last_review_date"].median(), inplace=True)
# Reseñas
Tourist_Accommodation["reviews_per_month"].fillna(0, inplace=True)
review_columns = [
    "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness",
    "review_scores_checkin", "review_scores_communication", "review_scores_location",
    "review_scores_value"
]
Tourist_Accommodation[review_columns] = Tourist_Accommodation[review_columns].apply(lambda x: x.fillna(x.median()))
# Verificar si quedan nulos
print(Tourist_Accommodation.isnull().sum())
```

```python
# Realizamos cambios a enteros
Tourist_Accommodation['bathrooms'] = Tourist_Accommodation['bathrooms'].astype('int')
Tourist_Accommodation['bedrooms'] = Tourist_Accommodation['bedrooms'].astype('int')
Tourist_Accommodation['beds'] = Tourist_Accommodation['beds'].astype('int')
```

```python
# VALORES DUPLICADOS
# Primero, encontramos las filas duplicadas por 'apartment_id'
duplicates = Tourist_Accommodation[Tourist_Accommodation.duplicated(subset=['apartment_id'], keep=False)]

# Filtrar para ver si son exactos o parciales:
# 1. Duplicados exactos: Todas las columnas deben coincidir
duplicates_exact = duplicates[duplicates.duplicated(keep=False)]

# 2. Duplicados parciales: Al menos algunas columnas coinciden (por ejemplo, `apartment_id` y otra columna clave)
duplicates_partial = duplicates[~duplicates.duplicated(subset=['apartment_id', 'name'], keep=False)]

# Contar el número de duplicados exactos y parciales
exact_count = duplicates_exact.groupby('apartment_id').size()
partial_count = duplicates_partial.groupby('apartment_id').size()

# Mostrar resultados
print("Duplicados exactos por apartment_id:")
print(exact_count)

print("\nDuplicados parciales por apartment_id:")
print(partial_count)
```

```python
# Ver duplicados parciales con las columnas relevantes
duplicates_partial = Tourist_Accommodation[Tourist_Accommodation.duplicated(subset=['apartment_id', 'name'], keep=False)]

# Mostrar las filas duplicadas parciales para inspección
print(duplicates_partial)

# Para ver las diferencias entre las filas duplicadas parciales
duplicates_diff = duplicates_partial.groupby('apartment_id').apply(lambda group: group.drop_duplicates())

# Ver las diferencias entre las filas duplicadas parciales por apartment_id
print("\nDiferencias entre duplicados parciales:")
print(duplicates_diff)
```

```python
# Calculamos el rango intercuartílico (IQR)
Q1 = Tourist_Accommodation['price'].quantile(0.25)
Q3 = Tourist_Accommodation['price'].quantile(0.75)
IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR # Calculamos el límite superior

outliers_precio_mas_alto = Tourist_Accommodation[Tourist_Accommodation['price'] > upper_limit] # Filtramos los outliers de precios más altos
print(outliers_precio_mas_alto[['name', 'price', 'bedrooms', 'bathrooms', 'accommodates', 'beds']])
```

```python
# En este caso no hay outliers inferiores en el conjunto de datos. Es decir, no hay propiedades con precios que sean tan bajos en comparación con el resto de los precios, según el criterio que definimos usando el rango intercuartílico (IQR).
lower_limit = Q1 - 1.5 * IQR # Calculamos el límite inferior
outliers_precio_mas_bajo = Tourist_Accommodation[Tourist_Accommodation['price'] < lower_limit]# Filtrar los outliers de precios más bajos
print(outliers_precio_mas_bajo[['name', 'price', 'bedrooms', 'bathrooms', 'accommodates', 'beds']])
```
